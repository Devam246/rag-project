```python
# STEP 1 : Reading the document
!pip install langchain pypdf
!pip install langchain-community
from langchain_community.document_loaders import PyPDFLoader

# Define the path to your PDF
pdf_path = "/syllabusCSE.pdf"

# Create a loader instance
loader = PyPDFLoader(pdf_path)

# Load the document
# This returns a list of Document objects, one for each page
pages = loader.load()

# You can check the number of pages loaded
print(f"Loaded {len(pages)} pages from the PDF.")

# You can access the content of a specific page
# The text is stored in the 'page_content' attribute
print("\n--- Content of First Page ---")
print(pages[0].page_content) # mera cmt , this cmd only outputs the first page of the doc readed, waise saare 55 pgs
#are read by PyPDFLoader that we imported from the langchain community
```

```python
# STEP 2 : Making टोकन from the read document/pdf
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    length_function=len
)

chunks = text_splitter.split_documents(pages)

# --- Check the output ---
print(f"Split the {len(pages)} pages into {len(chunks)} chunks.")

# Print the content of the first chunk
print("\n--- Content of First Chunk ---")
print(chunks[0].page_content)

# Print the number of characters in the first chunk
print(f"\nNumber of characters in the first chunk: {len(chunks[0].page_content)}")
```

```python
#STEP 3 : वेक्टर embeddings and storing them in the vector db
# First, make sure you have the required libraries installed
!pip install faiss-cpu sentence-transformers

from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

# --- Use the "fast" Hugging Face model to create embeddings ---
model_name = "all-MiniLM-L6-v2"
embeddings_model = HuggingFaceEmbeddings(model_name=model_name)

# --- Create the vector store using FAISS ---
# This single line creates embeddings for all your chunks and stores them
print("Creating vector store...")
vector_store = FAISS.from_documents(chunks, embeddings_model)
print("Vector store created successfully!")
```

```python
# First, install the necessary library for the LLM
!pip install transformers torch

from langchain_community.llms import HuggingFacePipeline
from langchain.chains import RetrievalQA
import torch

# (We assume your 'vector_store' from the previous step is ready)

# 1. Create a retriever from the vector store
retriever = vector_store.as_retriever()

# 2. Set up the LLM (we'll use a free model from Hugging Face)
# This may take a minute or two to download the model
llm = HuggingFacePipeline.from_model_id(
    model_id="google/flan-t5-base",
    task="text2text-generation",
    model_kwargs={"temperature": 0.6, "max_length": 500},
)

# 3. Create the QA chain
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True  # Optional: To see which chunks were used
)

# 4. Ask a question!
query = "what is subject code for DBMS"
result = qa_chain.invoke({"query": query})

# Print the answer
print("Answer:", result["result"])
```
